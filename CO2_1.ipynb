```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from prophet import Prophet
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM

# ==========================================
# 1. 데이터 준비 (df_all 사용)
# ==========================================
df = df_all.copy()

# '월' 내 숫자만 추출 (예: "1월" → 1)
df['월_num'] = df['월'].str.replace('월','').astype(int)

# 차량_배출량 → 숫자로 변환
df['차량_배출량'] = pd.to_numeric(df['차량_배출량'], errors='coerce')

# 날짜 생성
df['날짜'] = pd.to_datetime(df['연도'].astype(str) + '-' + df['월_num'].astype(str) + '-01')
df = df.set_index('날짜')

# 시계열 데이터 설정
ts = df['차량_배출량'].astype(float)


forecast_steps = 6
future_index = pd.date_range(ts.index[-1] + pd.offsets.MonthBegin(),
                             periods=forecast_steps, freq='MS')

# ==========================================
# 2. Baseline Naive
# ==========================================
baseline_fc = np.repeat(ts.iloc[-1], forecast_steps)

# ==========================================
# 3. ARIMA
# ==========================================
arima_fit = ARIMA(ts, order=(1,1,1)).fit()
arima_fc = arima_fit.forecast(steps=forecast_steps)

# ==========================================
# 4. SARIMA (계절성 제거 안정화)
# ==========================================
sarima_fit = SARIMAX(ts, order=(1,1,1), seasonal_order=(0,0,0,12)).fit()
sarima_fc = sarima_fit.forecast(steps=forecast_steps)

# ==========================================
# 5. Prophet
# ==========================================
df_prophet = ts.reset_index()
df_prophet.columns = ["ds", "y"]

prophet_model = Prophet()
prophet_model.fit(df_prophet)

future = prophet_model.make_future_dataframe(periods=forecast_steps, freq="MS")
prophet_pred = prophet_model.predict(future)
prophet_fc = prophet_pred[['ds', 'yhat']].tail(forecast_steps)
prophet_fc.set_index("ds", inplace=True)

# ==========================================
# 6. LSTM
# ==========================================
values = ts.values.reshape(-1,1)
scaler = MinMaxScaler()
scaled = scaler.fit_transform(values)

# 시퀀스 생성
def create_window(data, size=3):
    X, y = [], []
    for i in range(len(data)-size):
        X.append(data[i:i+size])
        y.append(data[i+size])
    return np.array(X), np.array(y)

window = 3
X, y = create_window(scaled, window)
X = X.reshape((X.shape[0], window, 1))

model_lstm = Sequential([
    LSTM(32, activation='relu', input_shape=(window,1)),
    Dense(1)
])
model_lstm.compile(optimizer='adam', loss='mse')
model_lstm.fit(X, y, epochs=300, verbose=0)

cur = scaled[-window:]
lstm_preds = []
for _ in range(forecast_steps):
    pred = model_lstm.predict(cur.reshape(1, window, 1), verbose=0)[0][0]
    lstm_preds.append(pred)
    cur = np.append(cur[1:], pred)

lstm_fc = scaler.inverse_transform(np.array(lstm_preds).reshape(-1,1)).flatten()

# ==========================================
# 7. 결과 정리
# ==========================================
forecast_df = pd.DataFrame({
    "날짜": future_index,
    "Baseline": baseline_fc,
    "ARIMA": arima_fc.values,
    "SARIMA": sarima_fc.values,
    "Prophet": prophet_fc['yhat'].values,
    "LSTM": lstm_fc
}).round(2)

print("\n📌 6개월 예측 결과 비교표")
print(forecast_df)

# ==========================================
# 8. 모델별 시각화
# ==========================================
models = {
    "Baseline": baseline_fc,
    "ARIMA": arima_fc.values,
    "SARIMA": sarima_fc.values,
    "Prophet": prophet_fc['yhat'].values,
    "LSTM": lstm_fc
}

for name, pred in models.items():
    plt.figure(figsize=(8,4))
    plt.plot(ts.index, ts.values, marker='o', label="Actual")
    plt.plot(future_index, pred, marker='o', label=name)
    plt.title(f"{name} Forecast - 차량 CO₂ 배출량")
    plt.xlabel("Month"); plt.ylabel("CO₂ Emission (kg)")
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from prophet import Prophet
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM

# ========================
# 1. 데이터 준비
# ========================
df = df_seoul_monthly.copy()
df['월'] = pd.to_datetime(df['월'])
df = df.set_index('월')

ts_2024 = df.loc["2024"]["승용_자가용_CO2_kg"].astype(float)

past_steps = 12
future_steps = 12
past_index = pd.date_range("2023-01-01", periods=past_steps, freq="MS")

# ========================
# Baseline
# ========================
baseline_back = np.repeat(ts_2024.iloc[0], past_steps)

# ========================
# ARIMA
# ========================
rev = ts_2024[::-1]
arima_fit = ARIMA(rev, order=(1,1,1)).fit()
arima_back = arima_fit.forecast(steps=past_steps).values[::-1]

# ========================
# SARIMA
# ========================
sarima_fit = SARIMAX(rev, order=(1,1,1), seasonal_order=(0,0,0,12)).fit()
sarima_back = sarima_fit.forecast(steps=past_steps).values[::-1]

# ========================
# Prophet (과거 12개월 강제 생성)
# ========================
prophet_df = ts_2024.reset_index()
prophet_df.columns = ["ds", "y"]
prophet_model = Prophet()
prophet_model.fit(prophet_df)

future_dates = pd.date_range("2023-01-01", "2024-12-01", freq="MS")
future_df = pd.DataFrame({"ds": future_dates})
prophet_pred = prophet_model.predict(future_df)

prophet_back = prophet_pred[prophet_pred['ds'] < "2024-01-01"]['yhat'].values
prophet_back = prophet_back[-12:]  # 정확히 12개월만 유지

# ========================
# LSTM
# ========================
values = ts_2024.values.reshape(-1,1)
scaler = MinMaxScaler()
scaled = scaler.fit_transform(values)

rev_scaled = scaled[::-1]
window = 3

def create_window(data, size):
    X, y = [], []
    for i in range(len(data)-size):
        X.append(data[i:i+size])
        y.append(data[i+size])
    return np.array(X), np.array(y)

X, y = create_window(rev_scaled, window)
X = X.reshape((X.shape[0], window, 1))

model = Sequential([
    LSTM(32, activation='relu', input_shape=(window,1)),
    Dense(1)
])
model.compile(optimizer='adam', loss='mse')
model.fit(X, y, epochs=200, verbose=0)

cur = rev_scaled[:window]
preds = []
for _ in range(past_steps):
    p = model.predict(cur.reshape(1, window, 1), verbose=0)[0][0]
    preds.append(p)
    cur = np.append(p, cur[:-1])

lstm_back = scaler.inverse_transform(np.array(preds).reshape(-1,1)).flatten()[::-1]

# ========================
# 결과 테이블
# ========================
backcast_df = pd.DataFrame(index=past_index)
backcast_df["Baseline"] = baseline_back
backcast_df["ARIMA"] = arima_back
backcast_df["SARIMA"] = sarima_back
backcast_df["Prophet"] = prophet_back
backcast_df["LSTM"] = lstm_back
backcast_df = backcast_df.round(2)

print("\n📌 2023년 역예측 결과")
print(backcast_df)

# ========================
# 📊 시각화: 2023(예측) + 2024(실측)
# ========================
plt.figure(figsize=(14,6))

for col in backcast_df.columns:
    plt.plot(backcast_df.index, backcast_df[col], linestyle='--', alpha=0.6)

plt.plot(ts_2024.index, ts_2024.values, linewidth=3, marker='o', color='black', label="2024 Actual")

plt.axvline(x=ts_2024.index[0], color='gray', linestyle=':', linewidth=2)

plt.title("2023 Backcasted + 2024 Measured CO₂ Emissions")
plt.xlabel("Month"); plt.ylabel("CO₂ Emission (kg)")
plt.xticks(rotation=45)
plt.legend(backcast_df.columns.tolist() + ["2024 Actual"])
plt.tight_layout()
plt.show()
```

구조적 일관성 검증

```python
def structural_gap(backcast, actual_2024):
    # 2023 마지막 예측 - 2024 첫 실측
    return abs(backcast[-1] - actual_2024[0])

gaps = []
for name in models.keys():
    gap = structural_gap(backcast_df[name].values, ts_2024.values)
    gaps.append([name, gap])

gap_df = pd.DataFrame(gaps, columns=["Model", "Gap(Dec2023→Jan2024)"])
gap_df = gap_df.sort_values("Gap(Dec2023→Jan2024)").reset_index(drop=True)
gap_df
```

트렌드 강도 비교

```python
def trend_strength(series):
    diff = np.diff(series)
    return np.var(diff)

trends = []
for name in models.keys():
    back_var = trend_strength(backcast_df[name].values)
    real_var = trend_strength(ts_2024.values)
    ratio = abs(back_var - real_var) / real_var  # 0에 가까울수록 좋음
    trends.append([name, back_var, ratio])

trend_df = pd.DataFrame(trends,
                        columns=["Model", "Backcast Var", "Var Difference Ratio"])
trend_df = trend_df.sort_values("Var Difference Ratio").reset_index(drop=True)
trend_df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# ======================================
# 1) 전체 데이터 구성 (2023 역예측 + 2024 실측)
# ======================================
full_ts = pd.concat([backcast_df["ARIMA"], ts_2024])
full_ts.index = pd.date_range("2023-01-01", periods=len(full_ts), freq="MS")

# ======================================
# 2) ARIMA 학습
# ======================================
arima_model = ARIMA(full_ts.values, order=(1,1,1)).fit()

# ======================================
# 3) 2025년 12개월 예측
# ======================================
forecast_steps = 12
arima_fc = arima_model.forecast(steps=forecast_steps)

future_dates = pd.date_range("2025-01-01", periods=forecast_steps, freq="MS")
arima_fc = pd.Series(arima_fc, index=future_dates)

print("\n📌 ARIMA 2025년 예측 결과")
print(arima_fc)

# ======================================
# 4) 시각화
# ======================================
plt.figure(figsize=(12,6))
plt.plot(full_ts.index, full_ts.values, marker='o', label='2023~2024 (Backcast+Actual)')
plt.plot(arima_fc.index, arima_fc.values, marker='o', label='ARIMA Forecast 2025')

plt.title("CO₂ 배출량 예측 (2023~2025, ARIMA 기반)")
plt.xlabel("Month"); plt.ylabel("CO₂ Emission (kg)")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
